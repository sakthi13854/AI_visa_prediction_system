```python
def assign_processing_center(state):
    east = ['NY','NJ','MA','PA','VA','MD','DC']
    central = ['TX','IL','OH','MI']
    west = ['CA','WA','OR','AZ']

    if state in east:
        return 'New York Service Center'
    elif state in central:
        return 'Texas Service Center'
    elif state in west:
        return 'California Service Center'
    else:
        return np.random.choice([
            'California Service Center',
            'Texas Service Center',
            'New York Service Center'
        ])

df['processing_center'] = df['WORKSITE_STATE'].apply(assign_processing_center)
```
