# Feature1 : service office specific avg processing time

```python
country_avg = df.groupby('SERVICE_OFFICE')['PROCESSING_DAYS'].mean()

```

```python
plt.figure(figsize=(10,6))
country_avg.sort_values(ascending=False).head(15).plot(kind='bar')
plt.title("service office Specific Avg Processing Time")
plt.xticks(rotation=60)
plt.show()
```

```python
visa_avg = df.groupby('VISA_CLASS')['PROCESSING_DAYS'].mean()

plt.figure(figsize=(10,6))
visa_avg.sort_values().plot(kind='bar')
plt.title("Average Processing Days by Visa Class")
plt.xticks(rotation=45)
plt.show()
```
